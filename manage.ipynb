```python
def display_3_lines(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:  # Menggunakan utf-8
            lines = file.readlines()
            for i, line in enumerate(lines[:3]):
                print(f"Line {i+1}: {line.strip()}")
    except FileNotFoundError:
        print(f"File '{filename}' not found.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
# Ganti 'nama_file.txt' dengan nama file yang ingin Anda baca
file_name = 'BindingDB_All_202309.tsv'
display_3_lines(file_name)
```
